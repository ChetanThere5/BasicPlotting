# Basic Plotting in Python using Matplotlib, Pandas and Seaborn

```python
#Import matplotlib
import matplotlib.pyplot as plt
#Below line defines to draw plot in notebook itself
%matplotlib inline
```

### Point

```python
plt.plot(7,2,'.');
```

```python
#Drawing red star point 
plt.plot(7,2,'*',c='r'); #semicolon avoids printing information at begining of figure
```

### Line Plot

```python
plt.plot([1,7,4]);
```

```python
#Plot dashed line with labels and title
x = [2,4,6]
y = [3,7,2]
plt.plot(x,y,'--o')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Line Plot');
```

```python
# Drawing two line plots in the same figure

linear = [i for i in range(1,5)]
exponential = [i**2 for i in linear]
plt.plot(linear,'-o',label='linear')
plt.plot(exponential,'-o',label='exponential')
plt.xlabel('entities')
plt.ylabel('growth')
plt.title('Growth Comparison')
plt.legend();
```

```python
# Fill color between two plots
plt.plot(linear,'-o',label='linear')
plt.plot(exponential,'-o',label='exponential')
plt.xlabel('entities')
plt.ylabel('growth')
plt.title('Growth Comparison')
plt.legend();
plt.gca().fill_between(range(len(linear)),linear,exponential,facecolor='blue',alpha=0.25);
```

### Bar Plot

```python
x = [20,40,60,80,100]
y = [11,33,55,77,99]
start = list(range(len(x)))
year=['2011','2012','2013','2014','2015']
plt.bar(start,x,width=0.3,label='item A')
start = [i+0.3 for i in start]
plt.bar(start,y,width=0.3,label='item B')
start = [i-0.2 for i in start]
plt.xticks(start,year)
plt.tick_params(bottom='off')
plt.xlabel('Year')
plt.ylabel('Sells')
plt.title('Sells of Item A and B')
plt.legend();
```

```python
#Vertical bar plot

x = [20,40,60,80,100]
y = [11,33,55,77,99]
start = list(range(len(x)))
year=['2011','2012','2013','2014','2015']
plt.bar(start,x,width=0.3)
plt.bar(start,y,width=0.3,bottom=x,color='c')
plt.xticks(start,year)
plt.tick_params(bottom='off')
plt.xlabel('Year')
plt.ylabel('Sells')
plt.title('Sells of Item A and B')
plt.legend(['item A', 'item B'],title='Legend',loc=2);  #another way to define legend, loc defines position of legend
```

```python
#Horizontal bar plot

x = [20,40,60,80,100]
y = [11,33,55,77,99]
start = list(range(len(x)))
year=['2011','2012','2013','2014','2015']
plt.barh(start,x,height=0.5)
plt.barh(start,y,height=0.5,left=x,color='c')
plt.yticks(start,year)
plt.tick_params(left='off')
plt.ylabel('Year')
plt.xlabel('Sells')
plt.title('Sells of Item A and B')
plt.legend(['item A', 'item B'],title='Legend',loc=4);
```

### Histogram


```python
#Import numpy : numpy supports high dimensional arrays and has many mathematical functions 
import numpy as np
normal_dist = list(np.random.normal(size=1000))
plt.hist(normal_dist,bins=50,color='blue',alpha=0.72); #aplha defines intensity of color
```

### Scatter Plot

```python
#scatter plot

x = np.random.randint(0,10,size=5)
y = np.random.randint(0,10,size=5)
x2 = np.random.randint(0,10,size=5)
y2 = np.random.randint(0,10,size=5)

plt.scatter(x,y,label='flowers')
plt.scatter(x2,y2,label='fruits',marker='x', color='r')
plt.xlabel('length')
plt.ylabel('width')
plt.title('Flower-Fruit')
plt.legend();
```

### Pie Chart

```python
#Pie Chart
x = np.random.randint(1,10,size=4)
y = ['office','exercise','sleep','study']
plt.pie(x,labels=y,explode=(0,0.1,0,0),startangle=90,autopct='%1.1f',shadow=True);
```

### Box plot

```python
plt.boxplot([linear,exponential])
plt.xticks([1,2],['linear','exponential'])
plt.title('Box Plot');
```

### Subplot

```python
plt.subplot(1,3,1)
plt.plot(linear,'--o')
plt.subplot(1,3,2)
plt.plot(exponential,'--o');
plt.subplot(1,3,3)
plt.plot(linear,'--o')
plt.plot(exponential,'--o');
```

### Plotting using pandas

```python
#Import pandas : library for data manipulation and analysis
import pandas as pd
normal_dist = np.random.normal(size=10)
random_dist = np.random.random(size=10)
# Create DataFrame
df = pd.DataFrame({'normal':normal_dist,'random':random_dist})
print(df.head(5))
```

```python
#line plot using DataFrame
df.plot();
```

```python
#Line plot using Series
df_normal = df['normal']
df_normal.plot();
```

```python
#Bar chart 
df.plot.bar();
```

```python
#Scatter plot 
df.plot('normal','random',kind='scatter');
```

```python
#kde plot
df.plot.kde();
```

```python
# Vertical bar chart 
df['normal'].plot.bar();
df['random'].plot.bar(bottom=df['normal'],color='c');
```

```python
#Box Plot
df.boxplot();
```

### Plotting using Seaborn

```python
#Import seaborn : library for statistical data visualization based on matplotlib
import seaborn as sns
normal_dist = np.random.normal(size=1000)
random_dist = np.random.random(size=1000)
df = pd.DataFrame({'normal':normal_dist,'random':random_dist})
```

```python
#scatter plot
sns.jointplot(df['normal'],df['random']);
```

```python
#kde plot
sns.jointplot(df['normal'],df['random'],kind='kde');
```

```python
#Import dataset from sklearn : library for machine learning
from sklearn import datasets
iris = datasets.load_iris()
features = iris.feature_names
X = pd.DataFrame(iris.data,columns=features)
y = pd.DataFrame(iris.target,columns=['target'])
iris_df = pd.concat([X,y],axis=1)
print(iris_df.head())
```

```python
sns.pairplot(iris_df,hue='target');
```
